# Make summary match panels

```python
# Select component
component = 'C3'
target = h_matrix.loc[component, :]
target.name = 'KRAS Component {}'.format(component)

# Select features
features = [
    [
        'Mutation',
        gct.read_gct('../data/mutation__gene_x_ccle_cellline.gct'),
        'high',
        'binary',
        '../results/match_{}_and_Mutation.match.txt'.format(component),
        [
            'KRAS_MUT',
            'APC_MUT',
            'CTNNB1_MUT',
        ],
        [
            'KRAS',
            'APC',
            'CTNNB1',
        ],
    ],
    [
        'Gene Set',
        gct.read_gct('../data/gene_set__gene_set_x_ccle_cellline.gct'),
        'high',
        'continuous',
        '../results/match_{}_and_Gene Set.match.txt'.format(component),
        [
            'SINGH_KRAS_DEPENDENCY_SIGNATURE_',
        ],
        [
            'KRAS Dependency',
        ],
    ],
    [
        'Gene Dependency (Achilles)',
        gct.read_gct('../data/achilles__gene_x_ccle_cellline.gct'),
        'low',
        'continuous',
        '../results/match_{}_and_Gene Dependency (Achilles).match.txt'.format(
            component),
        [
            'KRAS',
            'CTNNB1',
        ],
        [
            'KRAS',
            'CTNNB1',
        ],
    ],
]

# Make match summary panel
match_panel.make_summary_match_panel(
    target,
    features,
    title='Selected Features for {}'.format(target.name),
    file_path='../results/{}.summary_match_panel.pdf'.format(target.name))
```
