```python
from environment import *

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

### 2. Read the feature datasets

```python
mut_cna_df = ccal.read_gct('../data/ccle_mut_cna.gct')
gene_dependency_df = ccal.read_gct('../data/ccle_gene_dependency.gct')
gene_expression_df = ccal.read_gct('../data/rnaseq.v3.NO_HAEM.gct')
pathway_expression_df = ccal.read_gct('../data/ccle_pathway_expression_all.gct')
regulator_df = ccal.read_gct('../data/ccle_regulator.gct')
protein_expression_df = ccal.read_gct('../data/ccle_protein_expression.gct')
tissue_df = ccal.read_gct('../data/ccle_tissue.gct')
drug_sensitivity_df = ccal.read_gct('../data/ccle_drug_sensitivity.gct')
prom_methyl_df = ccal.read_gct('../data/ccle_prom_DNA_methyl.gct')
metabolites_df = ccal.read_gct('../data/ccle_metabolites.dat.gct')
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 3. Read oncogenic components and global clustering labels
<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 3.1 Read H matrix with all the KRAS components

```python
h_matrix = ccal.read_gct('../results/nmf/matrices/nmf_k9_h.gct')
```

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 3.2 Read the global clustering labels and select state labels for the k=12 oncogenic states

```python
global_clusterings = ccal.read_gct('../results/global/clusterings/clusterings.gct')
global_sample_labels = global_clusterings.ix[15, :]
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
#### 3.3 Display global Onco-GPS map

```python
component_names = ['C1 ERBB3/PI3K', 'C3 RAS/WNT/PI3K', 'C9 KRAS/AP1', 'C8 MYC',  'C6 BRAF/MAPK' , 
                                         'C7 TNF/NFkB',   'C5 HNF1/PAX8', 'C2 MYC/E2F', 'C4 EMT']

colors = ['#993300',  # Brown
                '#FFD700',  # Gold
                '#E74C3C',  # Red
                '#8B008B',  # Dark magenta  
                '#B0E0E6',  # Powder blue
                '#4169E1',  # Royal blue
                '#DC143C', # Crimson  
                '#669999',   # Green
                '#F4BD60',  # Light Sandy brown
                '#2E8B57',  # Sea green
                '#4b0082',  # indigo #000080',  # Naby blue 
                '#D2691E',  # Chocolate
                '#FA8072',  # Salmon
                '#DA70D6', # Orchid  
                '#90EE90']  # Light green 
```

```python
ccal.oncogps.make_oncogps(training_h = h_matrix, 
                                                  training_states = global_sample_labels, 
                                                  title = 'Global Onco-GPS Map',
                                                  component_markersize = 24,
                                                  component_fontsize = 22,
                                                  component_markerfacecolor='#000080',
                                                  sample_markersize = 18,
                                                  mds_seed = 1234,
                                                  std_max = 2,
                                                  state_colors = colors, 
                                                  power = 2.5,
                                                  kde_bandwidth_factor=1.25,
                                                  informational_mds = False,
                                                  component_names = component_names)
                                                  # filepath = DIR_RESULT + '/Global_Onco-GPS.pdf')
```

#### 3.4 Display the KRAS and RAC1 mutant cancers

```python
for i, alias in [('KRAS_MUT','KRAS mutants'),
                        ('RAC1_MUT', 'RAC1 mutants'),
                        ('RAC1_AMP', 'RAC1 amplifications')]:
    
    annotation = mut_cna_df.ix[i, :]
    annotation.name = alias

    ccal.oncogps.make_oncogps(training_h = h_matrix, 
                                                  training_states = global_sample_labels, 
                                                  annotation_type = 'binary',
                                                  training_annotation=annotation,
                                                  title=annotation.name,
                                                  power = 2.5,
                                                  component_markersize = 24,
                                                  component_fontsize = 22,
                                                  component_markerfacecolor='#000080',
                                                  sample_markersize = 18,
                                                  mds_seed = 1234,
                                                  std_max = 2,
                                                  informational_mds = False,
                                                  kde_bandwidth_factor=1.25,
                                                  state_colors = colors,
                                                  component_names = component_names)
                                                  #filepath = DIR_RESULT + 'Global_Onco-GPS_RAC1_Feature_{}.pdf'.format(annotation.name))
    
    ccal.plot_violin_box_or_bar(global_sample_labels, annotation.ix[global_sample_labels.index], colors=colors,
                                                  violin_or_box='box', figure_size=(8, 3))
```

#### 3.4 Display  only the RAC1 mutant  cancers

```python
RAC1 = tissue_df.columns[mut_cna_df.ix['RAC1_MUT', :].astype(bool)]
KRAS = tissue_df.columns[mut_cna_df.ix['KRAS_MUT', :].astype(bool)]
RAC1_h_matrix = h_matrix.ix[:, h_matrix.columns & RAC1]
RAC1_sample_labels = global_clusterings.ix[12, h_matrix.columns & RAC1]

RAC1_KRAS_wt_h_matrix = h_matrix.ix[:, h_matrix.columns & (RAC1 & (1 - KRAS))]
#RAC1_KRAS_wt _sample_labels = global_clusterings.ix[12, h_matrix.columns & (RAC1 & (1 - KRAS))]
```

```python
ccal.oncogps.make_oncogps(training_h=h_matrix, 
                              training_states=global_sample_labels, 
                              testing_h=RAC1_h_matrix,
                              testing_states=RAC1_sample_labels ,
                              testing_h_normalization='using_training_h',
                              power = 2.5,
                              title = 'RAC1 mutant cancers',
                              component_markersize = 24,
                              component_fontsize = 22,
                              component_markerfacecolor='#000080',
                              sample_markersize = 18,
                              sample_markersize = 25,
                              mds_seed = 1234,
                              std_max = 2,
                              informational_mds = False,
                              kde_bandwidth_factor=1.25,
                              component_names = component_names,
                              state_colors = colors,
                              sample_name_size=9, 
                              sample_name_color='blue')
                              #filepath = DIR_RESULT + 'Onco-GPS_RAC1_Feature_{}.pdf'.format(annotation.name))    
```

#### 3.5 Display selected features on the RAC1 mutant cancer samples

```python
for i, alias in [('AIGNER_ZEB1_TARGETS', 'ZEB1 Targets'),
                        ('HINATA_NFKB_TARGETS_FIBROBLAST_UP', 'NFKB Activation')]:
    
    annotation = pathway_expression_df.ix[i, :]
    annotation.name = alias

    ccal.oncogps.make_oncogps(training_h=h_matrix, 
                              training_states=global_sample_labels, 
                              testing_h=RAC1_h_matrix,
                              testing_states=RAC1_sample_labels ,
                              testing_h_normalization='using_training_h',
                              training_annotation=annotation,
                              title=annotation.name,
                              power = 2.25,
                              component_markersize = 24,
                              component_fontsize = 22,
                              component_markerfacecolor='#000080',
                              sample_markersize = 18,
                              sample_markersize = 25,
                              mds_seed = 1234,
                              std_max = 2,
                              kde_bandwidth_factor=1.25,
                              informational_mds = False,
                              component_names = component_names,
                              state_colors = colors)
                              #filepath = DIR_RESULT + 'Liver_Onco-GPS_RAC1_Feature_{}.pdf'.format(annotation.name))     
    
    ccal.plot_violin_box_or_bar(global_sample_labels, annotation.ix[global_sample_labels.index], colors=colors,
                                                  violin_or_box='box', figure_size=(10,3.5))
    
```

#### Find Features Associated with RAC1 mutations

```python

features_bundle = [
   ['Tissue Types', tissue_df, 'binary', False, 0, [], []],
   ['Protein Expression', protein_expression_df, 'continuous', False, 0, [], []],
   ['Mutations and CNA', mut_cna_df, 'binary', False, 0, [], []],
   ['Gene Dependency', gene_dependency_df, 'continuous', True, 0, [], []],
   ['Gene Expression', gene_expression_df, 'continuous', False, 0, [], []],
   ['Pathway Expression', pathway_expression_df, 'continuous', False, 0, [], []],
   ['Regulators', regulator_df, 'continuous', False, 0, [], []],
   ['Drug Sensitivity', drug_sensitivity_df, 'continuous', True, 0, [], []],
   ['Promoter Methylation', prom_methyl_df, 'continuous', True, 0, [], []],
   ['Metabolites', metabolites_df, 'continuous', True, 0, [], []]
]

target = mut_cna_df.ix['RAC1_MUT', :]
target.name = 'RAC1 mutations'
for f in features_bundle:
    print('RAC1 mut vs {}'.format(f[0]))
    ccal.association.make_association_panel(target = target, 
                                                                features = f[1], 
                                                                features_type = f[2], 
                                                                n_features = 20,          
                                                                n_permutations = 0,
                                                                title= 'RAC1 vs {}'.format(f[0]),  
                                                                features_ascending = f[3],
                                                                filepath_prefix= '../results/RAC1/RAC1_MUT_vs_{}'.format(f[0]),
                                                                random_seed = 12345)     
```
