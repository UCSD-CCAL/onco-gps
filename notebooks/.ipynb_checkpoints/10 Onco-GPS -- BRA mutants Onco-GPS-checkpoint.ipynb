### 1. Set up notebook and import Computational Cancer Analysis Library ([CCAL](https://github.com/KwatME/ccal))

```python
import sys
import os
import time
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.rcParams['figure.figsize'] = (8, 5)
mpl.rcParams['figure.max_open_warning'] = 100
HOME_DIR = os.environ['HOME']
sys.path.insert(0, os.path.join(HOME_DIR, 'ccal'))
import ccal
import warnings
warnings.filterwarnings("ignore")
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 2. Set parameters, data and result directories
These are the directories where the output and intermediate files will be stored

```python
DATA_DIR = HOME_DIR + '/OncoGPS_Analysis2/data'
RESULTS_DIR = HOME_DIR + '/OncoGPS_Analysis2/results'

BRAF_RESULTS_DIR = HOME_DIR + '/OncoGPS_Analysis/results/BRAF'
MUT_CNA_DATASET = DATA_DIR + '/ccle_mut_cna.gct'
GENE_DEPENDENCY_DATASET = DATA_DIR +  '/ccle_gene_dependency.gct'
GENE_EXPRESSION_DATASET = DATA_DIR + '/rnaseq.v3.NO_HAEM.gct'
PATHWAY_EXPRESSION_DATASET = DATA_DIR + '/ccle_pathway_expression_all.gct'
REGULATOR_DATASET = DATA_DIR + '/ccle_regulator.gct'
PROTEIN_EXPRESSION_DATASET = DATA_DIR + '/ccle_protein_expression.gct'
TISSUE_DATASET = DATA_DIR + '/ccle_tissue.gct'
DRUG_SENSITIVITY_DATASET = DATA_DIR + '/ccle_drug_sensitivity.gct'
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 3. Load the Feature Datasets

```python
mut_cna_df = ccal.read_gct(MUT_CNA_DATASET)
gene_dependency_df = ccal.read_gct(GENE_DEPENDENCY_DATASET)
gene_expression_df = ccal.read_gct(GENE_EXPRESSION_DATASET)
pathway_expression_df = ccal.read_gct(PATHWAY_EXPRESSION_DATASET)
regulator_df = ccal.read_gct(REGULATOR_DATASET)
protein_expression_df = ccal.read_gct(PROTEIN_EXPRESSION_DATASET)
tissue_df = ccal.read_gct(TISSUE_DATASET)
drug_sensitivity_df = ccal.read_gct(DRUG_SENSITIVITY_DATASET)
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 4. Read oncogenic components and select the subset BRAF mut samples

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 4.1 Select BRAF mut samples 

```python
blue, green, red, purple, black, white = '#4292C6', '#41AB5D', '#EF3B2C', '#807DBA', '#000000', '#FFFFFF'
braf_mutants = mut_cna_df.columns[mut_cna_df.ix['BRAF_MUT', :].astype(bool)]
h_matrix = ccal.read_gct(RESULTS_DIR + '/nmf/matrices/nmf_k9_h.gct')
braf_mutant_h_matrix = h_matrix.ix[:, h_matrix.columns & braf_mutants]
```

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 4.2 Display BRAF mut samples

```python
ccal.plot_heatmap(dataframe = braf_mutant_h_matrix, cluster=True,
                              xlabel = 'Sample', 
                              ylabel = 'KRAS Component', 
                              xticklabels = False)
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 5. Compare the KRAS components against each other in the BRAF mutant samples

```python
ccal.association.make_comparison_panel(matrix1 = braf_mutant_h_matrix, 
                                                                             matrix2 = braf_mutant_h_matrix,
                                                                             axis = 1, 
                                                                             matrix1_label = 'Component', 
                                                                             matrix2_label = 'Component')
```

As described in the analysis overview, we can see in the association matrix above that in those samples (BRAF mut) the components cluster into three major groups: 6-2, 8-9-3-1, 5-7-4.   This means that we can use 3 representative components (C6, C7 and C3) to define BRAF mut relevant oncogenic states and Onco-*GPS* map.

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 6. Select the BRAF mutants samples in the subset of C6, C7 and C3 components.

```python
braf_h_matrix = braf_mutant_h_matrix.ix[[ 7, 3, 6], :]
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 6. Define the Oncogenic States
This function performs the clustering of the KRAS $H$ matrix to define oncogenic states.

```python
components = pd.DataFrame(index=['Vertex 1', 'Vertex 2', 'Vertex 3'], columns=['x', 'y'])
components.iloc[0, :] = [0.5, np.sqrt(3)/2]
components.iloc[1, :] = [1, 0]
components.iloc[2, :] = [0, 0]

braf_dist_matrix, braf_clusterings, braf_cophen_coeff = ccal.oncogps.define_states(matrix = braf_h_matrix, 
                                                                                                                                                      ks = range(2, 11), 
                                                                                                                                                      max_std = 3.25, 
                                                                                                                                                      directory_path = BRAF_RESULTS_DIR)
```

As we can see in the cophenetic coefficient plot above there is a solution for k=3 states. We will select this number and make a heatmal and the correpsonding Onco-*GPS* below.

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 7. Make a heatmap of the selected (k=3) states

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 7.1 Select the state labels for the k=3 oncogenic states

```python
binary_labels = ccal.oncogps.define_binary_state_labels(braf_clusterings, 3, state_relabeling = [1, 2, 3])
ccal.write_gct(binary_labels, BRAF_RESULTS_DIR +  '/clusterings/clusterings_k3.gct')
braf_sample_labels = binary_labels.ix[4,:]
```

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 7.2 Produce heatmap of BRAF mutant samples and their memebrship in each state (this is Fig S11 in the article)

```python
ccal.plot_heatmap(dataframe = braf_h_matrix, 
                                  annotation_colors = [blue, green, red],
                                  data_type = 'continuous', 
                                  normalization_method = '-0-', 
                                  normalization_axis = 1,
                                  column_annotation = braf_sample_labels,
                                  title = 'H-Matrix Clustering',
                                  xlabel = 'Sample', 
                                  ylabel = 'Component', 
                                  xticklabels = False)
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 8. Generate the BRAF mut Onco-GPS map 
This function generates the Onco-*GPS* that we featured in the article and that we will use in the next notebooks. This is Fig 7A in the article.

```python
ccal.oncogps.make_oncogps(training_h = braf_h_matrix, 
                                                    training_states = braf_sample_labels, 
                                                    title = 'BRAF mut Onco-GPS Map',
                                                    components = components,  
                                                    colors= [blue, green, red],
                                                    component_names = ["C7", "C3", "C6"],
                                                    std_max = 2,
                                                    filepath = BRAF_RESULTS_DIR + '/BRAF_Onco-GPS.pdf')
```
