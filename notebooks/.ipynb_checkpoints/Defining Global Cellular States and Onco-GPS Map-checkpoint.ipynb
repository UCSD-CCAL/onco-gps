### 1. Set up notebook and import Computational Cancer Analysis Library ([CCAL](https://github.com/KwatME/ccal))

```python
from environment import *

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 2. Read the oncogenic components 
<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 2.1 Read H matrix with all the KRAS components

```python
h_matrix = ccal.read_gct('../results/nmf/matrices/nmf_k9_h.gct')
```

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 2.2 Display samples

```python
ccal.plot_heatmap(h_matrix, 
                  cluster=True,
                  xlabel='Sample', 
                  ylabel='KRAS Component', 
                  xticklabels=False)
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 3. Define the Oncogenic States

Cluster all the samples in the refrence dataset using all the 9 components. Based the Consensus-Clustering-Cophenetic-Correlation Coefficients Vs. K plot below, we choose a solution with large number of states to have increased resolution. We chose k=17.

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 4. Make a heatmap of the selected (k=10) states
<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 4.1 Select state labels for the k=10 oncogenic states

```python
global_clusterings = ccal.read_gct('../results/global/clusterings/clusterings.gct')
global_sample_labels = global_clusterings.ix[15, :]
```

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 4.2 Produce heatmap of KRAS mutant samples and their membership in each state (this is the top of Fig 6A)

```python
component_names = ['C1 ERBB3/PI3K', 'C3 RAS/WNT/PI3K', 'C9 KRAS/AP1', 'C8 MYC',  'C6 BRAF/MAPK' , 
                                         'C7 TNF/NFkB',   'C5 HNF1/PAX8', 'C2 MYC/E2F', 'C4 EMT']

colors = ['#993300',  # Brown
                '#FDD700',  # Gold
                '#E74C3C',  # Red
                '#8B008B',  # Dark magenta  
                '#B0E0E6',  # Powder blue
                '#4169E1',  # Royal blue
                '#DC143C', # Crimson  
                '#669999',   # Green
                '#F4BD60',  # Light Sandy brown
                '#2E8B57',  # Sea green
                '#4b0082',  # indigo #000080',  # Naby blue 
                '#D2691E',  # Chocolate
                '#FA8072',  # Salmon
                '#DA70D6', # Orchid  
                '#90EE90']  # Light green 
ccal.plot_heatmap(h_matrix, 
                  data_type='continuous',
                  annotation_colors = colors,
                  normalization_axis=1,
                  normalization_method='-0-',
                  column_annotation=global_sample_labels,
                  title='H-Matrix Clustering by State',
                  xlabel='Sample', 
                  ylabel='Component', 
                  xticklabels=False)
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 5. Generate global Onco-GPS map 

This is Fig S1 in the article.

```python
ccal.oncogps.make_oncogps(training_h = h_matrix, 
                                                  training_states = global_sample_labels, 
                                                  title = 'Global Onco-GPS Map',
                                                  power = 2.5,
                                                  component_markersize = 24,
                                                  component_fontsize = 22,
                                                  component_markerfacecolor='#000080',
                                                  sample_markersize = 18,
                                                  mds_seed = 1234,
                                                  std_max = 2,
                                                  state_colors = colors,
                                                  kde_bandwidth_factor=1.25,
                                                  informational_mds = False)
                                                  #filepath = DIR_RESULT + '/Global_Onco-GPS.pdf')

h_matrix.index = [c[:2] for c in component_names]
ccal.oncogps.make_oncogps_in_3d(h_matrix, global_sample_labels, '../results/Global_Onco-GPS_3D', state_colors=colors)
```
