### 3 Read the $H$  matrix and select state labels
<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 3.1 Read H matrix with all the KRAS components and the clustering labels and select state labels for the k=4 oncogenic states

```python
kras_mutants = mutation_df.columns[mutation_df.ix['KRAS_MUT', :].astype(bool)]
h_matrix = ccal.read_gct('../results/nmf/matrices/nmf_k9_h.gct')
kras_mutant_h_matrix = h_matrix.ix[:, h_matrix.columns & kras_mutants]
kras_h_matrix = kras_mutant_h_matrix.ix[[7, 2, 1], :]

binary_labels = ccal.read_gct('../results/clusterings/clusterings_k4.gct')
kras_sample_labels = binary_labels.ix[5, :]
```

<hr style="border: none; border-bottom: 3px solid #88BBEE;">
### 4. Display component amplitudes as features (this is Fig S7 in the article)

```python
for i, alias in [(1, 'C1'),(2, 'C2'),(3, 'C3'), (4, 'C4'),(5, 'C5'),
                        (6, 'C6'),(7, 'C7'),(8, 'C8'),(9, 'C9')]:
    
    annotation = kras_mutant_h_matrix.ix[i, kras_h_matrix.columns]
    annotation.name = alias
    
    ccal.oncogps.make_oncogps(training_h=kras_h_matrix, 
                              training_states=kras_sample_labels, 
                              annotation=annotation,
                              title=annotation.name,
                              equilateral=True,
                              colors=[purple, blue, red, green],
                              power = 1.3,
                              kde_bandwidths_factor=0.8,
                              legend_markersize=25,
                              contour_linecolor=light_blue,
                              component_names=["C7","C2","C1"],
                              std_max=2)
                              # filepath=join(DIR_RESULT, 'KRAS_Onco-GPS_Feature_{}.pdf'.format(annotation.name))) 
```

### 6. Produce Figure 7F

```python
test_samples = ['NCIH1792_LUNG', 'SW620_LARGE_INTESTINE', 'NCIH23_LUNG', 'NCIH2122_LUNG', 'LOVO_LARGE_INTESTINE',
                             'DV90_LUNG']

test_h_matrix = kras_h_matrix.ix[:, test_samples]

test_sample_labels = binary_labels.ix[5, test_samples]

phenotype = [1, 1, 1, 0, 0, 0]
annotation = ccal.support.d1.make_series(phenotype, index=test_h_matrix.columns, name='selection')
```

```python
test_samples = ['NCIH1792_LUNG', 'SW620_LARGE_INTESTINE', 'NCIH23_LUNG', 'NCIH2122_LUNG', 'LOVO_LARGE_INTESTINE',
                             'DV90_LUNG']

test_h_matrix = kras_h_matrix.ix[:, test_samples]

test_sample_labels = binary_labels.ix[5, test_samples]

phenotype = [0, 0, 0, 1, 1, 1]
annotation = d1.make_series(phenotype, index=test_h_matrix.columns, name='selection')

ccal.oncogps.make_oncogps(training_h=kras_h_matrix, 
                              training_states=kras_sample_labels,
                              testing_h=test_h_matrix,
                              testing_states=test_sample_labels ,
                              annotation=annotation,
                              title =annotation.name,
                              annotation_type='binary',
                              testing_h_normalization='using_training_h',
                              equilateral=True,
                              legend_markersize=25,
                              sample_markersize=32,
                              colors=[purple, blue, red, green],
                              # sample_name_color = '#000000',
                              sample_name_size=14,
                              kde_bandwidths_factor=0.7,
                              component_names=["C7","C2","C1"],
                              std_max=2)
                              # filepath=join(DIR_RESULT,'KRAS_Onco-GPS.pdf'))
```

### 7. Pan-Can analysis

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 7.1 Read Pancan datasets

```python
#PANCAN_MUT_CNA = DIR_DATA + '/PANCAN12.mutation.whitelist.v2.1.gct'
PANCAN_MUT_CNA = '../data/PANCAN12.mutation.whitelist.v2.1.out.gct'

#PANCAN_GENE_EXPRESSION = DIR_DATA + '/PanTCGA_expression_profile.out.v2.SUBSET.gct'
PANCAN_GENE_EXPRESSION = '../data/PanTCGA_expression_profile.out.gct'

# PANCAN_PATHWAY_EXPRESSION = DIR_DATA + '/PanTCGA_expression_profile.out.v2.MSIGDB_plus_paper_complementary_signatures.gct'
PANCAN_PATHWAY_EXPRESSION = '../data/PanTCGA_expression_profile.out.MSIGDB_plus_paper_complementary_signatures.gct'

#PANCAN_PROTEIN_EXPRESSION =  '../data/PANCAN_TCGA_RPPA.gct'
PANCAN_PROTEIN_EXPRESSION =  '../data/RPPA_TCGA.out.gct'

#PANCAN_H_MATRIX =  DIR_DATA + '/PanTCGA_expression_profile.out.v2.SUBSET.LUNG_KRAS_MUT.H_proj.v1.gct'
PANCAN_H_MATRIX  = '../data/PanTCGA_expression_profile.out.v2.LUNG_KRAS_MUT.H_proj.v1.gct'

pancan_mut_cna_df = ccal.read_gct(PANCAN_MUT_CNA)
pancan_gene_expression_df = ccal.read_gct(PANCAN_GENE_EXPRESSION)
pancan_pathway_expression_df = ccal.read_gct(PANCAN_PATHWAY_EXPRESSION)
pancan_protein_expression_df = ccal.read_gct(PANCAN_PROTEIN_EXPRESSION)
pancan_h_matrix = ccal.read_gct(PANCAN_H_MATRIX)
```

<hr style="border: none; border-bottom: 1px solid #88BBEE;">
#### 7.2 Association analysis for KRAS components C3 and C7 in pancan datasets

```python
features_bundle = [
    #['Protein Expression', pancan_protein_expression_df, 'continuous', False, 0, [], []],
    #['Mutations and CNA', pancan_mut_cna_df, 'binary', False, 0, [], []],
    #['Gene Expression', pancan_gene_expression_df, 'continuous', False, 0, [], []]
    ['Pathway Expression', pancan_pathway_expression_df, 'continuous', False, 0, [], []]
]

for component in [2, 6]:   # Components C3 and C7 are entries 2 and 6 (base zero)
    target = pancan_pathway_expression_df.ix[component, :]
    target.name = 'KRAS Component C{}'.format(component + 1)
    for f in features_bundle:
        print('{} vs {}'.format(target.name, f[0]))
        ccal.association.make_association_panel(target = target, 
                                    features = f[1], 
                                    features_type = f[2], 
                                    n_features = 20,          
                                    n_permutations = 5,
                                    title= '{} vs {}'.format(target.name, f[0]),  
                                    features_ascending = f[3],
                                    filepath_prefix= '../results/C{}'.format(component + 1) + '/PANCAN_kras_component_C{}_vs_{}'.format(component + 1,  f[0]),
                                    random_seed = 12345)   
```
